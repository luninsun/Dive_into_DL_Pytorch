```python
import torch
import torchvision
from torch import nn, optim

from PIL import Image
import d2lzh_pytorch as d2l
```

```python
d2l.set_figsize()
img = Image.open('img/cat.jpeg')
d2l.plt.imshow(img)
```

### 水平翻转

```python
d2l.apply(img, torchvision.transforms.RandomHorizontalFlip(), scale = 2.5)
```

### 垂直翻转

```python
d2l.apply(img, torchvision.transforms.RandomVerticalFlip(), scale = 2.5)
```

### 随机裁剪

```python
shape_aug = torchvision.transforms.RandomResizedCrop(200, scale = (0.1, 1), ratio = (0.5, 2))
d2l.apply(img, shape_aug, scale = 2.5)
```

### 变换颜色之亮度

```python
d2l.apply(img, torchvision.transforms.ColorJitter(brightness = 0.5), scale = 2.5)
```

### 变换颜色之色调

```python
d2l.apply(img, torchvision.transforms.ColorJitter(hue = 0.5), scale = 2.5)
```

### 变换颜色之对比度

```python
d2l.apply(img, torchvision.transforms.ColorJitter(contrast = 0.5), scale = 2.5)
```

### 变换颜色之饱和度

```python
d2l.apply(img, torchvision.transforms.ColorJitter(saturation = 0.5), scale = 2.5)
```

### 变换颜色之亮度、色调、对比度、饱和度

```python
color_aug = torchvision.transforms.ColorJitter(brightness=0.5, contrast=0.5, saturation=0.5, hue=0.5)
d2l.apply(img, color_aug, scale = 2.5)
```

### 🌟🌟叠加多个图像增广方法（实际使用）

```python
shape_aug1 = torchvision.transforms.RandomHorizontalFlip()
shape_aug2 = torchvision.transforms.RandomVerticalFlip()
shape_aug3 = torchvision.transforms.RandomResizedCrop(200, scale = (0.6, 1), ratio = (0.5, 2))
color_aug = torchvision.transforms.ColorJitter(brightness=0.5, contrast=0.5, saturation=0.5, hue=0.5)
augs = torchvision.transforms.Compose([
    shape_aug1,
    shape_aug1,
    shape_aug3,
    color_aug
])

d2l.apply(img, augs, scale = 2.5)
```
